```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import math
```

```python
frame0 = cv2.imread('../imgs/charger_0.jpg',0)
frame1 = cv2.imread('../imgs/charger_1.jpg',0)

orig0 = frame0
orig1 = frame1

f = plt.figure()
f.add_subplot(1,2, 1)

plt.imshow(orig0, cmap="gray")
plt.title('Frame 0'), plt.xticks([]), plt.yticks([])

f.add_subplot(1,2, 2)

plt.imshow(orig1, cmap="gray")
plt.title('Frame 1'), plt.xticks([]), plt.yticks([])

plt.show(block=True)
```

```python
frame0 = cv2.resize(frame0,(400,300))
frame1 = cv2.resize(frame1,(400,300))

f = plt.figure()
f.add_subplot(1,2, 1)

plt.imshow(frame0, cmap="gray")
plt.title('Frame 0'), plt.xticks([]), plt.yticks([])

f.add_subplot(1,2, 2)

plt.imshow(frame1, cmap="gray")
plt.title('Frame 1'), plt.xticks([]), plt.yticks([])

plt.show(block=True)
```

```python
absdiff = cv2.absdiff(frame0,frame1)
plt.imshow(absdiff, cmap='gray', vmin=0)
plt.title('absdiff'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
ret,thresh = cv2.threshold(absdiff,127,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(thresh, cmap='gray', vmin=0)
plt.title('diff_thresh'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
diff = frame1 - frame0

plt.imshow(diff, cmap='gray', vmin=0)
plt.title('diff'), plt.xticks([]), plt.yticks([])
plt.show()
```
